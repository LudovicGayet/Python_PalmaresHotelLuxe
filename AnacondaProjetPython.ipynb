```python
import pandas as pd

```

```python
def lectureFichier(fichier):
    return pd.read_csv(fichier, sep=";", header=0)

critere = lectureFichier("critere.csv")
data = lectureFichier("hotel.csv")

```

```python
def norm1(crit, h):
    if type(h) == str :
        data[crit] = data[crit].apply(lambda x : x.replace(",", "."))
    data[crit] = data[crit].apply(lambda x : float(x))
    h = float(h)
    maxi = max(data[crit])
    mini = min(data[crit])
    if critere[crit][0] == 'max':
        return float((h - mini) / (maxi - mini))
    if critere[crit][0] == 'min':
        return float((h - maxi) / (mini - maxi))

def norm2(crit, h):
    if type(h) == str :
        data[crit] = data[crit].apply(lambda x : x.replace(",", "."))
    data[crit] = data[crit].apply(lambda x : float(x))
    h = float(h)
    maxi = max(data[crit])
    mini = min(data[crit])
    if critere[crit][0] == 'max':
        return float(h / maxi)
    if critere[crit][0] == 'min':
        return float(1 - (h/maxi))

dataNorm1 = pd.DataFrame(index=data.index, columns=data.columns)
dataNorm1.nom_hotel = data.nom_hotel
dataNorm1.pays = data.pays

dataNorm2 = pd.DataFrame(index=data.index, columns=data.columns)
dataNorm2.nom_hotel = data.nom_hotel
dataNorm2.pays = data.pays

for i in range(len(data.columns)):
    if i not in [0,6]:
        for j in range(len(data.index)):
            dataNorm1.iloc[j].iloc[i] = norm1(data.columns[i], data.iloc[j].iloc[i])
            dataNorm2.iloc[j].iloc[i] = norm2(data.columns[i], data.iloc[j].iloc[i])

print(dataNorm1)
print(dataNorm2)
```
